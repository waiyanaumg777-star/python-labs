# Activity 1: Identify the Components

### 1.1. Inputs
* Quiz 1 mark
* Quiz 2 mark
* Quiz 3 mark
* Continuation choice ('y' or 'n')

### 1.2. Process
* Sum the three quiz marks and divide by 3 to compute the average.
* Check if the average is greater than or equal to 50.
* Use a loop to repeat the program if the user enters 'y'.

### 1.3. Output
* The calculated average mark.
* A message showing a "Pass" or "Fail" status.

# Activity 2: Design the Algorithm

### 2.1. Pseudocode
```text
START
    SET choice = "Y"
    WHILE choice is "Y" or "y" DO
        INPUT quiz_1, quiz_2, quiz_3
        COMPUTE average = (quiz_1 + quiz_2 + quiz_3) / 3
        OUTPUT average
        
        IF average >= 50 THEN
            OUTPUT "Pass"
        ELSE
            OUTPUT "Fail"
        ENDIF
        
        INPUT choice ("Continue? Select Y/N: ")
    ENDWHILE
    OUTPUT "Program Ended"
END
```

### 2.2. Algorithm Structures
* Sequence: Requesting the inputs sequentially and computing the average score.
* Selection: Using an IF/ELSE conditional statement to determine a pass or fail.
* Iteration: Running a WHILE loop to continuously process additional marks.
